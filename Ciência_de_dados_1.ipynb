### 2. Amostra Sistemática

Como realizar amostragem sistemática (com padrões)

### Moda


```python
amodal = pd.DataFrame([1,2,3,4,5,6,7,8,9,1])

#mode() -> moda -> CRIA UMA TABELA MODA
#shape[0] -> conta todas as linhas de um dataFrame
#Pega o número de linhas do df 'amodal.mode' e compara com o nº de linhas do df 'amodal'
amodal.mode().shape[0]
if amodal.mode().shape[0] == amodal.shape[0]:
  print('amodal')
#elif -> else if
#Se o nº de linhas da tabela amodal.mode == 1, tem apenas uma moda
elif amodal.mode().shape[0] == 1:
  print('modal')
#E assim por diante
elif amodal.mode().shape[0] == 2:
  print('bimodal')
else:
  print('polimodal')
```

```python
lista = [1.63, 1.65, 1.72, 1.79, np.NaN, 1.82, np.NaN, 1.99, 2.00]
medianaImpar = pd.DataFrame(lista);

# O parâmetro NaN vem "true" padrão no método "median()". Desativando ele, passa a contar os NaN na execução do método
medianaImpar.median(skipna=False)

# O método "fillna(n)" substitui todos os valores NaN pelo número "n" passado como parâmetro
# Porém, depois de substituir os valores, automaticamente ordena o DataFrame
print(medianaImpar.fillna(0).median())
```

####Criando um gráfico de todas as colunas numéricas do DataFrame

## 3. Medidas de dispersão

### Amplitude

```python
amplitude = pd.DataFrame([4,6,79,1,64,4,33,4,9,6,4])
# A amplitude tira o valor máximo contido no array do valor mínimo contido no array
A = amplitude.max() - amplitude.min()
A
```

#### __Atividade 01:__ O setor de controle de qualidade de uma empresa seleciona ao acaso peças de um lote. Quando a amplitude das medidas dos diâmetros das peças ultrapassa 0,8 cm o lote é rejeitado. Considerando que em um lote foram encontrados os seguintes valores 2,1 cm; 2,0 cm; 2,2 cm; 2,9 cm; 2,4 cm, esse lote foi aprovado ou rejeitado?

```python
lim = 0.8

amplitude = pd.DataFrame([2.1,2,2.2,2.9,2.4])
A = amplitude.max() - amplitude.min()

# O método da amplitude (cálculo realizado acima) RETORNA UM ARRAY
  # Por isso o index [0] ao lado da amplitude A
if A[0] > lim:
  print(f'O lote teve amplitude {A[0]} e foi rejeitado!')
else:
  print(f'O lote teve amplitude {A[0]} e foi aprovado!')
```

#### __Atividade 02:__ Considere 4 prédios com alturas 460m, 800m, 300m e 400m. Qual a amplitude entre as alturas dos prédios?

```python
amplitude = pd.DataFrame([460,800,300,400])
A = (amplitude.max() - amplitude.min())[0]

print(f"A amplitude entre as alturas dos prédios é {A}")
```

#### __Atividade 03:__ Aplique o cálculo de amplitude para cada coluna numérica do DataFrame Titanic

```python
numericDF = df.select_dtypes(include=['int', 'float', 'bool'])

for column in numericDF:
  A = df[column].max() - df[column].min()
  print(f"A amplitude da coluna {column} é: {A}")

# Jeito mais fácil
amplitudes = df.max(numeric_only=True) - df.min(numeric_only=True)
amplitudes
```
